```python
import matplotlib.pyplot as plt
import numpy as np

m2ft = 3.28

ax = plt.figure().add_subplot(projection='3d')
ax.set_title("Knuckleball Trajectory")
ax.set_xlabel('x (ft)')
ax.set_ylabel('y (ft)')
ax.set_zlabel('z (ft)')

for deg in ["0.2", "30"]:
    file_name = f"knuckleball.wy{deg.replace('.', '')}.json.dat"
    x = []
    y = []
    z = []
    with open(file_name) as f:
        for line in f:
            x_value, y_value, z_value, _, _, _ = line.split()
            x.append(float(x_value) * m2ft)
            y.append(float(y_value) * m2ft)
            z.append(float(z_value) * m2ft)

    ax.plot(x, y, z, label=f"{deg} rev/s")

# ax.view_init(elev=0, azim=-90, roll=0) # xz
ax.legend()
plt.show()
```
